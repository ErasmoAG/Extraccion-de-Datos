```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Carga desde un archivo csv
df = pd.read_csv('Euskadi.csv')
df
```

```python
#Verificamos informacion del dataframe
df.info()
```

**1er Punto: Los registros de 5 "host name" diferentes**

```python
#FILTRO POR OBJETO Y FILTRO Y
filtro1= df[(df["host_name"] == "Mikel") | (df["host_name"] == "Emy") | (df["host_name"] == "Maitane") | (df["host_name"] == "Concha") | (df["host_name"] == "Ainhoa")]
filtro1
```

**2do Punto: Los registros de los host que se hallan unido a Airnab "antes del a;o 2019"**

```python
#FILTRO POR COMPARACION
filtro2=df[df["host_since"] < "2019-01-01"]
filtro2
```

**3er Punto: Los registros de los host que se hallan unido a Airnab "antes del a;o 2019"**

```python
filtro3=df[df["host_response_time"] == "within a day"]
filtro3
```

**4to Punto: Los registros que tengan "al menos 5 accomodates"**

```python
filtro4=df[df["accommodates"] >= 5]
filtro4
```

**5to Punto: Los registros de los tipos de cuarto "Entire home" y ""Private room**

```python
filtro5=df[(df["room_type"] == "Entire home/apt") | (df["room_type"] == "Private room")]
filtro5
```

**6to Punto: Los registros que "a lo mucho cuenten con 4 camas"**

```python
filtro6=df[(df["beds"] <= 4)]
filtro6
```

**7mo Punto: Los registros que "tengan 1 ba;o" o que sean superhost" pero con la condicion de que "todos sean instant_bookable"**

```python
filtro7=df[((df["bathrooms"] == 1) | (df["host_is_superhost"] == "t")) & (df ["instant_bookable"] == "t")]
filtro7
```

**8vo Punto: Los registros cuyo "review_scores_cleanliness sea menor de 3 revies_scores_reting sea mayor de 3"**

```python
filtro8=df[(df["review_scores_cleanliness"] < 3) & (df ["review_scores_rating"] > 3)]
filtro8
```

**Creacion de los archivos separados por comas (punto csv)**

```python
#CONVERTIR ARCHIVO FILTRADO a CSV
filtro1.to_csv("5_host_name_diferentes.csv")
filtro2.to_csv("host_antes_de_2019.csv")
filtro3.to_csv("respuesta_a_mas_tardar_1_dia.csv")
filtro4.to_csv("al_menos_5_accomodates.csv")
filtro5.to_csv("entire_home_y_private_room.csv")
filtro6.to_csv("a_lo_mucho_cuenten_con_4_camas.csv")
filtro7.to_csv("1_bano_o_que_sean_superhost_todos_sean_instant_brokable.csv")
filtro8.to_csv("review_scores_cleanliss_menor_de_3_y_review_scores_raiting_mayor_a_3.csv")
```
