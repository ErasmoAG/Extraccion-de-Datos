```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Carga desde un archivo .xlsx sin indice
df = pd.read_excel('Detalle precios y productos fabricados 2022.xlsx')
df
```

```python
#Verificamos informacion del dataframe
df.info()
```

**PRIMER FILTRO: FILTRO POR OBJETO**

```python
#FILTRO POR OBJETO
filtro1= df[(df["NOMBRE_VENDEDOR"] == "LETICIA RAMIREZ HERNANDEZ")]
filtro1
```

**SEGUNDO FILTRO: FILTRO POR FILAS**

```python
#FILTRO POR FILAS
filtro2 = df.iloc[3:11,: ]                     #FILAS CONSECUTIVAS
#filtro2 = df.iloc [ [2, 5] , :]               #FILAS NO CONSECUTIVAS
filtro2
```

**TERCER FILTRO: FILTRO POR COLUMNAS**

```python
#FILTRO POR COLUMNAS 
#filtro3 = df.iloc[ : , 3: 12 ]    #COLUMNAS consecutivas (Columnas con la informacion del vendedor y el cleitne, hasta el calculo de costos)
filtro3 = df.iloc [ : , [2,3,5]]    #COLUMNAS NO CONSECUTIVAS (Columnas especificas para ver al cliente, al vendedor y su clave)
filtro3
```

```python
#CARGA 
df1 = pd.read_excel('Detalle precios y productos fabricados 2022.xlsx', index_col=2)
df1
```

**CUARTO FILTRO: FILTRO FILAS VS COLUMNAS**

```python
#PASO 2: APLICAMOS FILTRO DE FILAS VS COLUMNAS
filtro4=df1.loc[["LETICIA RAMIREZ HERNANDEZ","DIEGO ARMANDO LEON DURAN"], ["MARGEN_TOTAL_CALCULADO"]]
filtro4
```

**QUINTO FILTRO: FILTRO POR CABECERA**

```python
#FILTRO POR CABECERA
filtro5=df.head()
filtro5
```

**SEXTO FILTRO: FILTRO POR COLA**

```python
#FILTRO POR COLA
filtro6=df.tail()
filtro6
```

**SEPTIMO FILTRO: FILTRO POR COMPARACION**

```python
#FILTRO POR COMPARACION
filtro7=df[df["COSTO_TOTAL_CALCULADO"] > 6000]
filtro7
```

**OCTAVO FILTRO: FILTRO PARA DOS CONDICIONES &**

```python
#FILTRO Y
filtro8=df[(df["NOMBRE_VENDEDOR"] == "LETICIA RAMIREZ HERNANDEZ" ) & (df["FECHA_DOC"] == "2022-12-22")]
filtro8
```

**NOVENO FILTRO: FILTRO POR UNA DE DOS CONDICIONES |**

```python
#FILTRO O
filtro9=df[(df["MARGEN_TOTAL_CALCULADO"] < 6000) | (df["FECHA_DOC"] == "2022-12-22")]
filtro9
```

**DECIMO FILTRO: FILTRO POR NINGUNA DE LAS DOS CONDICIONES NOT**

```python
#FILTRO NOT
filtro10=df[~(df["FECHA_DOC"] == "2022-12-22")]
filtro10
```

**CREACION DE LOS ARCHIVOS SEPARADOS POR COMAS (CSV)**

```python
#CONVERTIR ARCHIVO FILTRADO a CSV
filtro1.to_csv("Filtro_Objeto_Especifico.csv")
filtro2.to_csv("Filtro_Filas.csv")
filtro3.to_csv("Filas_VS_ColumnasColumnas.csv")
filtro4.to_csv("Filas_VS_ColumnasFilas_VS_Columnas.csv")
filtro5.to_csv("Filtro_por_Cabecera.csv")
filtro6.to_csv("Filtro_por_Cola.csv")
filtro7.to_csv("Filtro_por_Comparacion.csv")
filtro8.to_csv("Filtro_Y.csv")
filtro9.to_csv("Filtro_O.csv")
filtro10.to_csv("Filtro_NOT.csv")
```
